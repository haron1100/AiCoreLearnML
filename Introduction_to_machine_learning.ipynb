```
from sklearn import datasets #import the datasets from sklearn libary

X, Y = datasets.load_boston(return_X_y=True) #load the boston housing dataset

print('Design matrix shape:', X.shape) 
print('Labels shape:', Y.shape) 
```

```
from sklearn import linear_model #import linear models from sklearn
import numpy as np

model = linear_model.LinearRegression(normalize=True) #create linear model
model.coef_ = np.random.randn(13) #randomly initialise the scaling coefficients randomly
model.intercept_ = np.random.randn() #randomly initialise b randomly

y0_pred = model.predict(X[0].reshape(1,-1)) #make a prediction on the first data point
print('Prediction of median house price (in thousands) on first first town in', y0_pred)
```

```
from sklearn import metrics #import metrics from sklearn

Y_pred = model.predict(X) #use our model to make a prediction for each data point in our dataset
L = metrics.mean_squared_error(Y, Y_pred) #calculate the mean squared error between our predictions and the label
print('Mean squared error of model on this dataset is', L)
```

```
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(np.arange(len(Y)), Y_pred, label="Predictions")
plt.scatter(np.arange(len(Y)), Y, label="Labels")
plt.xlabel('Town number')
plt.ylabel('Median house price (in thousands)')
plt.legend()
plt.show()
```

```
model.fit(X, Y) #run gradient descent to optimise our models parameters to reduce the cost
Y_pred = model.predict(X) #use our trained model to make a prediction for each data point in our dataset

L = metrics.mean_squared_error(Y, Y_pred) #calculate the mean squared error between our predictions and the label
print('Mean squared error of the trained model on this dataset is', L)
```

```
plt.figure()
plt.scatter(np.arange(len(Y)), Y_pred, label="Predictions")
plt.scatter(np.arange(len(Y)), Y, label="Labels")
plt.xlabel('Town number')
plt.ylabel('Median house price (in thousands)')
plt.legend()
plt.show()
```
